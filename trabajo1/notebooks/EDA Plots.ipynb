```python
import pandas as pd
pd.set_option('display.max_columns', None)
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
s3_path = 'https://trusted-proyecto-integrador.s3.us-east-1.amazonaws.com/pmo_data/part-00000-eae00926-15e1-4e10-aced-0bb86568869a-c000.snappy.parquet'
```

```python
df = pd.read_parquet(s3_path)
```

```python
df.head()
```

### Distribucion: `Duration`

```python
plt.figure(figsize=(12, 5))
hist = df['duration'].hist(color = '#414141', alpha = 0.85)
plt.title('Histograma: Duration', loc = 'left',  fontsize = 16.5, fontweight = 'bold', color = '#414141');
plt.ylabel('Frecuencia', fontsize = 10, color = '#414141',labelpad = 15)
plt.xlabel('Duration', fontsize = 10, color = '#414141',labelpad = 15)
plt.yticks( fontsize = 10, color = '#414141');
plt.xticks( fontsize = 10, color = '#414141');
```

```python
df.select_dtypes(include = 'float').corr()
```

```python

corr_matrix = df.select_dtypes(include = 'float').corr()
for i in corr_matrix.columns:
    corr_matrix[i] = round(corr_matrix[i],3)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.set_theme(style="white")

plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap="coolwarm", vmin=-1, vmax=1, center=0, square=True, linewidths=.5)

plt.title("Matriz de Correlación")
plt.show()
```
